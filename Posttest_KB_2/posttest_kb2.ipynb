```python
# inisialisasi library pandas
import pandas as pd

# mengambil dataset di file csv
fd = pd.read_csv("winemag-data-130k-v2.csv");
fd
```

## 1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
# 1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

# Seleksi atribut yang bertipe angka
atribut_angka = fd.select_dtypes(include=['number'])

# Hitung statistik yang diinginkan
statistics = atribut_angka.describe()
# Tambahkan perhitungan kuartil Q1, Q2 (median), dan Q3
statistics.loc['Q1'] = atribut_angka.quantile(0.25)
statistics.loc['Q2'] = atribut_angka.median()
statistics.loc['Q3'] = atribut_angka.quantile(0.75)

print(statistics)
```

## 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
# 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

# Informasi jumlah record
jumlah_records = len(atribut_angka)

# Informasi jumlah attribute
jumlah_attributes = len(fd.columns)

# Informasi nama attribute
nama_attributes = fd.columns.tolist()

# Informasi jumlah attribute dengan tipe angka
atribut_angka = fd.select_dtypes(include=['number'])
jumlah_atribut_angka = len(atribut_angka.columns)

# Membuat dictionary
dataset_info = {
    'Jumlah Record': jumlah_records,
    'Jumlah Attribute': jumlah_attributes,
    'Nama Attribute': nama_attributes,
    'Jumlah Attribute dengan Tipe Angka': jumlah_atribut_angka
}

# Tampilkan isi dictionary
for key, value in dataset_info.items():
    print(f"{key}: {value}")
```

## 4. Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.

```python
# 4. Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.

# Menggunakan iloc untuk melakukan slicing
slice_iloc = fd.iloc[120:, 3:]

# Tampilkan hasil slice
print(slice_iloc)

```

```python
# Menggunakan .loc untuk slicing
slice_loc = fd.loc[120:, fd.columns[3:]]

# Tampilkan hasil slice
print(slice_loc)
```

## 5. Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

```python
# 5. Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

atribut_objek = fd.select_dtypes(include=['object'])


# Tampilkan jumlah label unik untuk setiap atribut objek
for column in atribut_objek.columns:
    unique_labels = fd[column].nunique()
    print(f"Jumlah label untuk '{column}': {unique_labels}")
```
