# GRAFIK BATANG
### Menampilkan 5 jenis anggur(varietas) termahal berdasarkan Harga

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset CSV ke dalam DataFrame
data = pd.read_csv('winemag-data-130k-v2.csv')

# Kelompokkan data berdasarkan jenis varietas dan hitung rata-rata harga
varietas_harga_rata = data.groupby('variety')['price'].mean()

# Urutkan hasil pengelompokan berdasarkan harga secara menurun
varietas_harga_rata_sorted = varietas_harga_rata.sort_values(ascending=False)

# Ambil 3 jenis varietas teratas
top_5_varieties = varietas_harga_rata_sorted.head(5)


```

```python
top_5_varieties
```

```python
# Buat diagram batang
plt.figure(figsize=(10, 6))
top_5_varieties.plot(kind='bar', color=['red','skyblue','green','yellow','purple'])
plt.title('Top 5 Jenis Varietas Termahal')
plt.xlabel('Varietas')
plt.ylabel('Harga Rata-Rata')
plt.xticks(rotation=45)
plt.show()
```

### Berdasarkan visualisasi diagram batang yang menggambarkan top 5 jenis anggur (varietas) termahal dari dataset winemag-data-130k-v2.csv, dapat disimpulkan hal-hal berikut:

### Varietas Termahal: Varietas anggur yang paling mahal dari dataset adalah "Ramisco" dengan harga rata-rata sebesar    495.000.000.
### Varietas Kedua: Setelah Ramisco, varietas anggur "Terrantez" menduduki peringkat kedua dalam hal harga rata-rata, dengan nilai sekitar 236.000.000.
### Varietas Ketiga: "Francisa" adalah varietas anggur termahal ketiga dengan harga rata-rata sekitar 160.000.000.
### Varietas Keempat: Di peringkat keempat, terdapat varietas anggur "Rosenmuskateller" dengan harga rata-rata sekitar 150.000.000.
### Varietas Kelima: Terakhir, varietas anggur "Malbec-Cabernet" menduduki peringkat kelima dengan harga rata-rata sekitar 113.333.333.

### Dengan demikian, visualisasi data ini memberikan informasi penting tentang jenis-jenis anggur yang memiliki harga rata-rata tertinggi dalam dataset, dengan Ramisco menjadi yang paling mahal di antara semuanya. # GRAFIK HISTOGRAM
### Menampilkan jumlah orang yang memberikan penilaian berdasarkan points

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset CSV ke dalam DataFrame
data = pd.read_csv('winemag-data-130k-v2.csv')

# Plot histogram untuk distribusi poin yang diberikan oleh taster
plt.figure(figsize=(10, 6))
plt.hist(data['points'], bins=20, edgecolor='k', alpha=0.7)
plt.xlabel('Poin')
plt.ylabel('Jumlah Penilaian')
plt.title('Distribusi Poin yang Diberikan oleh Taster')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Kesimpulan Berdasarkan Grafik Histogram

## Berdasarkan visualisasi menggunakan diagram histogram pada dataset "winemag-data-130k-v2.csv"untuk melihat poin yang diberikan oleh para taster, kita dapat membuat beberapa kesimpulan terkait sebaran kualitas anggur dalam data set ini:

## Distribusi Poin: Dari histogram, terlihat bahwa sebagian besar anggur dalam dataset ini menerima poin dalam rentang 85 hingga 90. Ini menunjukkan bahwa mayoritas anggur dalam dataset mendapatkan poin di kisaran tersebut, yang mungkin mengindikasikan bahwa sebagian besar anggur memiliki kualitas yang baik.
## Poin Paling Umum: Poin yang paling umum diberikan oleh para taster tampaknya berada di sekitar 87 hingga 88. Ini dapat dianggap sebagai "poin rata-rata" yang sering diberikan.

## Kesimpulannya, visualisasi menggunakan histogram membantu kita memahami sebaran kualitas anggur dalam dataset ini dan memberikan gambaran umum tentang bagaimana para taster memberikan poin kepada anggur. 

# GRAFIK PIE CHART
### Menampilkan 10 wilayah pendistribusi wine berdasarkan Provinsi

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset CSV ke dalam DataFrame
data = pd.read_csv('winemag-data-130k-v2.csv')

# Hitung jumlah anggur di setiap region_1
jumlah_per_region = data['province'].value_counts()

# Batasi jumlah region yang akan ditampilkan agar grafik tidak terlalu padat
n = 10  # Contoh: Tampilkan 10 region teratas
jumlah_per_region = jumlah_per_region.head(n)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(jumlah_per_region, labels=jumlah_per_region.index, autopct='%1.1f%%', startangle=180)
plt.title('10 Wilayah Pendistribusi wine terbanyak berdarkan provinsi')
plt.xlabel('Province')
plt.show()

```

# Kesimpulan Berdasarkan Diagram Pie Chart

### Berdasarkan visualisasi menggunakan diagram pie chart pada dataset "winemag-data-130k-v2.csv" untuk melihat 10 pendistribusi anggur terbesar berdasarkan provinsi (province), kita dapat membuat beberapa kesimpulan:

### California Mendominasi: Provinsi California adalah pendistribusi anggur terbesar dalam dataset, dengan 45.5% dari total. Ini menunjukkan bahwa sebagian besar anggur dalam dataset berasal dari California.
### Washington dan Bordeaux: Provinsi-provinsi lain yang memiliki kontribusi signifikan dalam distribusi anggur adalah Washington (10.8%) dan Bordeaux (7.5%). Meskipun California mendominasi, Washington dan Bordeaux masih memiliki andil yang signifikan dalam dataset.
### Tuscany, Oregon, dan Burgundy: Tuscany (7.4%), Oregon (6.7%), dan Burgundy (5.0%) juga memiliki andil yang cukup besar dalam distribusi anggur dalam dataset. Mereka mewakili berbagai daerah di seluruh dunia yang dikenal dengan produksi anggur berkualitas tinggi.
### Provinsi Lain: Provinsi-provinsi lain seperti Northern Spain (4.8%), Piedmont (4.7%), Mendoza Province (4.1%), dan Veneto (3.4%) juga menyumbang dalam distribusi anggur, meskipun dengan proporsi yang lebih kecil.

# GRAFIK HEATMAP

## Menghitung matriks korelasi dari atribut numerik points dan price

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca dataset CSV ke dalam DataFrame
data = pd.read_csv('winemag-data-130k-v2.csv')

# Pilih subset fitur yang ingin Anda analisis
subset_data = data[['points', 'price']]


# Hitung matriks korelasi
correlation_matrix = subset_data.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.xlabel('Poin (Points)')
plt.ylabel('Harga (Price)')
plt.title('Heatmap Hubungan antara Poin dan Harga Anggur')
plt.show()
```

# Kesimpulan Berdasarkan Diagram HeatMap

## Berdasarkan diagram heatmap yang menghitung matriks korelasi antara atribut "points" dan "price" dalam dataset "winemag-data-130k-v2.csv," kita dapat membuat beberapa kesimpulan:

## Korelasi Positif: Terlihat bahwa terdapat korelasi positif antara atribut "points" (penilaian anggur) dan "price" (harga anggur). Artinya, semakin tinggi poin yang diberikan oleh taster, cenderung harga anggur juga semakin tinggi. Ini dapat diinterpretasikan bahwa dalam dataset ini, anggur yang mendapatkan penilaian tinggi lebih cenderung memiliki harga yang lebih tinggi.
## Poin Rata-rata: Poin yang paling umum diberikan oleh para taster mungkin berkisar di sekitar nilai tertentu, tetapi harga anggur memiliki variasi yang signifikan di semua tingkat penilaian.
## Outlier: Terlihat ada beberapa anggur yang mendapatkan poin tinggi tetapi memiliki harga yang rendah (outlier negatif) dan sebaliknya, ada beberapa anggur yang mendapatkan poin rendah tetapi memiliki harga yang tinggi (outlier positif).

## Kesimpulannya, matriks korelasi dan heatmap ini memberikan gambaran visual tentang hubungan antara poin dan harga anggur dalam dataset ini.
